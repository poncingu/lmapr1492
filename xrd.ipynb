```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("6wCWi8tP1m0OLbQ4") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-12908")
```

# Diffraction des rayons X sur un cristal $ScAgSe_2$
Afin de d√©terminer les trois premiers pics du difractogramme li√© au cristal √©tudi√©, Le module pymatgen.analysis.diffraction.core est utilis√© pour un rayonnement incident Cu KŒ± avec une longeur d'onde Œª=1.54060 √Ö. Ces rayons X vont cr√©er des ph√©nom√®nes d'interf√©rences constructives ou destructives selon la direction. √Ä l'aide du mod√®le de diffraction et de certaines hypoth√®ses, il est possible de trouver la position des atomes dans le cristal. $2\theta$ est ici pris comme angle de r√©f√©rence. 



```python
#Cr√©ation d'un objet XRDCalculator
xrd = XRDCalculator(1.54060)

#Calcul du mod√®le de diffraction
pattern = xrd.get_pattern(structure)
```

# Premier pic

```python
#R√©cup√©ration de l'angle du premier pic
theta1 = pattern.x[0]

#R√©cup√©ration de l'intensit√© du premier pic
intensite1 = pattern.y[0]

#R√©cup√©ration des indices hkl du premier pic
hkl1 = pattern.hkls[0]

#R√©cup√©ration de la distance inter-plan
d_hkls1 = pattern.d_hkls[0]

print('Angle du premier pic : '+  '2ùúÉ = ' + str(theta1) + ' degr√©s')
print('Intensit√© du premier pic : ' + str(intensite1))
print('Indices hkl du premier pic :' + str(hkl1))
print("Valeur de d, la distance inter-plan :", d_hkls1)
```

# Deuxi√®me pic

```python
#R√©cup√©ration de l'angle du deuxi√®me pic
theta2 = pattern.x[1]

#R√©cup√©ration de l'intensit√© du deuxi√®me pic
intensite2 = pattern.y[1]

#R√©cup√©ration des indices hkl du deuxi√®me pic
hkl2 = pattern.hkls[1]

#R√©cup√©ration de la distance inter-plan
d_hkls2 = pattern.d_hkls[1]

print('Angle du deuxi√®me pic : '+  '2ùúÉ = ' + str(theta2) + ' degr√©s')
print('Intensit√© du deuxi√®me pic : ' + str(intensite2))
print('Indices hkl du deuxi√®me pic :' + str(hkl2))
print("Valeur de d, la distance inter-plan :", d_hkls2)
```

# Troisi√®me pic

```python
#R√©cup√©ration de l'angle du troisi√®me pic
theta3 = pattern.x[2]

#R√©cup√©ration de l'intensit√© du troisi√®me pic
intensite3 = pattern.y[2]

#R√©cup√©ration des indices hkl du troisi√®me pic
hkl3 = pattern.hkls[2]

#R√©cup√©ration de la distance inter-plan
d_hkls3 = pattern.d_hkls[2]

print('Angle du troisi√®me pic : '+  '2ùúÉ = '  + str(theta3) + ' degr√©s')
print('Intensit√© du troisi√®me pic : ' + str(intensite3))
print('Indices hkl du troisi√®me pic :' + str(hkl3))
print("Valeur de d, la distance inter-plan :", d_hkls3)
```
